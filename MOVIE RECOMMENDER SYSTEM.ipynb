```python
cv=TfidfVectorizer()
tfidf_matrix=cv.fit_transform(movies['genres'])
```

```python
movie_user = df.pivot_table(index='userId',columns='title',values='rating')
movie_user.head()
```

```python
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
indices=pd.Series(movies.index,index=movies['title'])
titles=movies['title']
def recommendations(title):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:21]
    movie_indices = [i[0] for i in sim_scores]
    return titles.iloc[movie_indices]
```

```python
recommendations('Toy Story (1995)')
```

```python
# Adjust your function based on the actual column names
def popularity_recommender(df, genre, min_rating_threshold, num_recommendations):
    genre_filtered = df[(df['genres'] == genre) & (df['rating'] >= min_rating_threshold)]
    sorted_movies = genre_filtered.groupby('title').agg({'rating': 'mean','rating': 'count'}).sort_values(by='rating', ascending=False)
    top_recommendations = sorted_movies.head(num_recommendations)
    return top_recommendations
```

```python
import ipywidgets as widgets
from IPython.display import display

# GUI for Popularity-based Recommender
genre_dropdown = widgets.Dropdown(
    options=df['genres'].unique(),
    value='Action',
    description='Genre:'
)

min_rating_slider = widgets.FloatSlider(
    value=3.5,
    min=0,
    max=5.0,
    step=0.1,
    description='Min Rating:'
)

num_recommendations_input = widgets.IntText(
    value=5,
    description='Num Recommendations:'
)

def on_button_click(b):
    genre = genre_dropdown.value
    min_rating = min_rating_slider.value
    num_recommendations = num_recommendations_input.value
    recommendations = popularity_recommender(df, genre, min_rating, num_recommendations)
    print("Popularity-based Recommendations:")
    print(recommendations)

button = widgets.Button(description="Get Recommendations")
button.on_click(on_button_click)

# Display the GUI
display(genre_dropdown, min_rating_slider, num_recommendations_input, button)
```
